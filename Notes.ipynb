```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xls = pd.ExcelFile('notes.xls')
MedicalServices = pd.read_excel(xls, 'MedicalServices')
SocialServices = pd.read_excel(xls, 'SocialServices')
ATN = pd.read_excel(xls, 'ATN')
GeneralSE = pd.read_excel(xls, 'GeneralSE') 
SocialSecurity = pd.read_excel(xls, 'SocialSecurity') 
EmpBusOwner = pd.read_excel(xls, 'EmpBusOwner')
Citizenship = pd.read_excel(xls, 'Citizenship')
Housing = pd.read_excel(xls, 'Housing')
EET  = pd.read_excel(xls, 'EET')
ConnDSRVR = pd.read_excel(xls, 'ConnDSRVR')
ETS  = pd.read_excel(xls, 'ETS')
WRC = pd.read_excel(xls, 'WRC')
```

## Service Connentions

#### Medical Service/Medicare

```python
a = MedicalServices.Participants.groupby(MedicalServices.Service).count()
```

```python
a
```

```python
colors = ['#5E2D79', '#CDB5CD']
explode = (0.1,0.1)
labels = ['Did not use the service','Used the Service']
```

```python
plt.figure(figsize=(15,14))
plt.subplot(2, 2,1)
plt.pie(a, colors=colors, startangle=90,frame=True, explode=explode,radius=2.5, autopct='%1.1f%%')
centre_circle = plt.Circle((0,0),1.2,color='black', fc='white',linewidth=0)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('off')
plt.tight_layout()
plt.title("Percentage of participants that used Medical Services / Medicare", color = 'purple')
plt.legend(labels=labels, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=3)
plt.plot()
```
